# Feature Engineering

```python
df[df['resolution']=='1920 x 1080']
# Re-adjusts resolution to 1920 x 1080
```

```python
total_pixel1=df['resolution'].str.split('x').str.get(0).str.strip()
# To break 'resolution' into substrings, extracts elements from each component at index 0 and removes any leading, and trailing whitespaces
total_pixel2=df['resolution'].str.split('x').str.get(1).str.strip()
# To break 'resolution' into substrings, extracts elements from each component at index 1 and removes any leading, and trailing whitespaces
```

```python
temp_df1=total_pixel1.reset_index()
temp_df2=total_pixel2.reset_index()
# Allows reseting the index back to the default 0, 1, 2 etc indexes
```

```python
temp_df1['resolution']=temp_df1['resolution'].astype('int')
temp_df2['resolution']=temp_df2['resolution'].astype('int')
# Casting the pandas object to integer
```

```python
temp_df=temp_df1['resolution']*temp_df2['resolution']
# Making a temporary set of values with 'resolution'
temp_df
```

```python
df.insert(18,'total_pixel',temp_df.values)
# Inserts column 'total_pixel' within temporary dataset at index 18
```

```python
df[df['total_pixel'] < 921600]
# Displays rows of 'total_pixel' which is less than 921600
```

```python
df[df['total_pixel']>=3686400]
# Displays rows of 'total_pixel' which is more than or equal to 3686400
```

```python
def display(data):
    if data<= 307200:
        return 'VGA'
    elif data<= 614400:
        return 'SVGA'
    elif data<= 409920:
        return 'FWVGA'
    elif data<= 921600:
        return 'HD'
    elif data<= 2073600:
        return 'FULL HD'
    elif data<= 3686400:
        return '2k'
    elif data<= 3440640:
        return '4k'
    elif data<= 4608000:
        return '5k'
    else:
        return '5k'
    # Check for data conditions and returns appropriate results
```

```python
temp=df['total_pixel'].apply(display)
# Updates the data with above results
```

```python
df.insert(19,'Display',temp.values)
# Inserts column 'Display' within temporary dataset at index 19
```

# Missing Values
Use KNN IMPUTER for missing values

```python
from sklearn.impute import KNNImputer
# Imports KNNImputer from sklearn
```

```python
tempdf=df.select_dtypes(include=['int64','float64']).drop(columns='price')
```

```python
impute = KNNImputer(n_neighbors=5)
```

```python
new=impute.fit_transform(tempdf)
```

```python
new_df = pd.DataFrame(new,columns=tempdf.columns)
# Creates new variable which manages columns
```

```python
new_df.isnull().sum()
# Displays the number of null values within columns
```

```python
new_df['price']=df['price']
# Copying the column 'price' from the old to new
```

```python
x_df=new_df.corr()['price'].reset_index()
```

```python
x=df.corr()['price'].reset_index()
```

```python
data=x.merge(x_df,on='index',suffixes=('_1', '_2'))
# Merges x_df on the index, with the suffixes 1 and 2 appended to any overlapping columns
```

```python
data
```

```python
fig, axes = plt.subplots(1,2,figsize=(18, 10))
sns.barplot(ax=axes[0],x='index',y='price_1',data=data)
plt.sca(axes[0])
plt.title('Before Imputing')
plt.xticks(rotation='vertical')
sns.barplot(ax=axes[1],x='index',y='price_2',data=data)
plt.sca(axes[1])
plt.title('After Imputing')
plt.xticks(rotation='vertical')
plt.show()
# Plots 2 bar plots
# First one showing a bar plot, plotted before imputing
# Second one showing a bar plot, plotted after imputing
```
